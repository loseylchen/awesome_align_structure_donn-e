```python
from bertalign import Bertalign

# Define the source and target texts
src = """Life is beautiful.
We must cherish every moment.
Happiness lies in simple things."""
tgt = """Das Leben ist schön.
Wir müssen jeden Moment schätzen.
Das Glück liegt in einfachen Dingen."""

# Initialize and align sentences
aligner = Bertalign(src, tgt)
aligner.align_sents()
aligner.print_sents()

```

```python
# Print results in Pharaoh format
alignments = aligner.result

# Corrected Pharaoh format
pharaoh_format = []
for alignment in alignments:
    if isinstance(alignment, list) and len(alignment) == 2:
        src_idx, tgt_idx = alignment
        pharaoh_format.append(f"{src_idx + 1}-{tgt_idx + 1}")

print("\n".join(pharaoh_format))
```

```python
from bertalign import Bertalign

# Define the source and target texts
src = """Life is beautiful.
We must cherish every moment.
Happiness lies in simple things."""
tgt = """Das Leben ist schön.
Wir müssen jeden Moment schätzen.
Das Glück liegt in einfachen Dingen."""

# Initialize and align sentences
aligner = Bertalign(src, tgt)
aligner.align_sents()

# Extract alignments
alignments = aligner.result

# Check if alignments are empty
if not alignments:
    print("No alignments were produced.")
else:
    # Prepare the desired output format
    result_lines = []
    for alignment in alignments:
        # Convert alignment pairs to the desired format
        if isinstance(alignment, list):
            formatted_alignment = " ".join([f"{src_idx}-{tgt_idx}" for src_idx, tgt_idx in alignment])
            result_lines.append(formatted_alignment)

    # Print the results in the requested format
    output = "\n".join(result_lines)
    print(output)
```
